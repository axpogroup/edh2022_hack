```python
import pandas as pd
import numpy as np
import json
```

```python
df_gbr = pd.read_csv("../output.csv")
df_gbr
```

```python
df_gxf = pd.read_csv("../dxf/volume_and_locations.csv", converters={
    "ground_center": json.loads,
    "bounding_box": json.loads
})
#df_gxf['ground_center'] = df_gxf['ground_center'].apply(makeArray, ',')
df_gxf
```

```python
xs_dxf = [p[0] for p in df_gxf.ground_center]
ys_dxf = [p[1] for p in df_gxf.ground_center]
xs_gbr = df_gbr.coordinate_lat
ys_gbr = df_gbr.coordinate_lon
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

fig = figure(figsize=(16, 6), dpi=80)
plt.scatter(xs_dxf, ys_dxf, s=2, marker='o', c='red')

plt.scatter(xs_gbr, ys_gbr, s=1, marker='x', c='green')

#plt.show()
plt.savefig("fig.svg", dpi=3000, format='svg')
```

```python
import plotly.express as px
import plotly.graph_objects as go
gbr = go.Scatter(x=xs_gbr, y=ys_gbr, mode='markers', marker=dict(size=3, color='green'))
dxf = go.Scatter(x=xs_dxf, y=ys_dxf, mode='markers', marker=dict(size=3, color='red'))

fig = go.Figure(data=[gbr, dxf])
fig.write_html("map.html")
fig.show()
```
